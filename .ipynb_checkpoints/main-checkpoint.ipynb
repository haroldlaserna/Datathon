## Definiendo función para pasar datos categóricos a ordinales

```python
def categorical_to_ordinal(data):
    to_transform  = data.to_numpy()
    transform_categorical = LabelEncoder()
    transform_categorical.fit(to_transform)
    return transform_categorical.transform(to_transform)
```

```python
from sklearn.model_selection import train_test_split
```

```python
datos.columns
```

```python
X = datos.drop(columns=["Reached.on.Time_Y.N", "ID"])
Y = datos["Reached.on.Time_Y.N"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
sns.pairplot(X_train, diag_kind="hist")
plt.show()
```

```python
sns.pairplot(X_test, diag_kind="hist")
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier

preprocesamiento_categorico = Pipeline(steps = [("onehot", OneHotEncoder())])

preprocesor = ColumnTransformer([
                                   (
                                    "categorico", preprocesamiento_categorico,
                                   [
                                    "Warehouse_block", "Gender", "Product_importance", "Mode_of_Shipment"                                     
                                    ])
                                ])

param_grid = {
    "MLP__hidden_layer_sizes": [(500,),(500,500)],
    "MLP__activation": ["identity", "logistic", "tanh", "relu"],
    "MLP__learning_rate": ["constant", "invscaling", "adaptive"],
}

pipe = Pipeline(steps = [("Preprocesamiento", preprocesor), ("standarscale", StandardScaler()),
                         ("MLP", MLPClassifier(verbose=False))],verbose = False)
grid = GridSearchCV(pipe,param_grid=param_grid,verbose=3,cv=2)
grid
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_estimator_
```

```python
grid.score(X_train, Y_train) #con
```

```python
grid.score(X_train, Y_train)
```

```python
print(classification_report(Y_train, grid.predict(X_train)))
```

```python
grid.score(X_test, Y_test)
```

```python
grid.score(X_test, Y_test)
```

```python
print(classification_report(Y_test, grid.predict(X_test)))
```
